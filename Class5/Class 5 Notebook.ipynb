```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data_master/master.csv', parse_dates = ['visit_date'])
```

```python
df['quarter'] = df['visit_date'].dt.quarter
```

```python
#grouping 
df.groupby('id').sum()
#sums everything - even lat long 
```

```python
#grouping by multiple things - make a list 
df.groupby(['day_of_week', 'holiday']).mean()
```

```python
#order matters here
df.groupby(['holiday','day_of_week']).mean()
```

**groupby() returns pandas dataframes, which means we can do a lot with these**

```python
#selecting columns works normally

df.groupby(['holiday','day_of_week']).mean()[['visitors', 'reserve_visitors']]
```

"holiday" and "day_of_week" are actually part of the index! Observe below: 

```python
df.groupby(['holiday','day_of_week']).mean()[['visitors', 'reserve_visitors']].index
```

If you want to put these back into the actual dataframe, you'd use reset_index

#### What if you want to use multiple aggregations? 

```python
#use agg function and pass in a list
df.groupby(['day_of_week', 'holiday']).agg(['count', np.mean, 'sum', 'std'])
```

```python
#you can also write your own functions to pass into agg
def is_missing(series):
    return pd.isnull(series).sum()
```

```python
df.groupby(['day_of_week', 'holiday']).agg(['count', np.mean, 'sum', 'std', is_missing])
```

**Because functions are first-class objects BABYYYYYYY**

This can be super handy to apply a custom function or model to different subsets in your dataset - for example, how did these assets perform in normal times vs. in recession times. 

```python
#a grouping is a weird intermediate object 
df.groupby('day_of_week')
```

Not every pandas method applies to this strange intermediate type of object
One workaround is to write a lambda function 

```python
#usig apply and a lambda function is a great way to get around this limitation

df.groupby('day_of_week').apply(lambda x: x.iloc[:10])
```

### The Magical Sliding Window

```python
df['visitors'].shift()
```

Gradient boosting has no notion of time built in to it, so we'll often need to create additional columns with, for example, yesterday's date, before we can use gradient boosting to model time series

```python
df['yesterday'] = df['visitors'].shift()
#this is going to create a "spillover" problem when we move to the "next" restaurant 
#the last observation for restaurant 1 is going to be the first observation for restaurant 2 
```

```python
#we can solve the above problem with groupby
df['yesterday'] = df.groupby('id')['visitors'].shift()
```

```python
df['yesterday'].isnull().sum()
```

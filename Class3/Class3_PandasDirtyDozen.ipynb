#### Why does df.loc[10:100] work? 

Because in this case, the index is actually a numeric value. if the index was not a numeric value, you'd have to use iloc, which looks things up by the index

```python
#start, stop, step applies to pandas indexing as well 
df.iloc[10:100, ::2]
```

What's more common is using some condition in order to create new values for your new column 

```python
#let's say we wanted to have a column that indicates weekend vs. not weekend 
expression_we = df['day_of_week'].isin(['Friday', 'Saturday', 'Sunday'])
```

#### What if you want to create an additional column based on my than one condition? 

Use select()

```python
df['visit_date'] = pd.to_datetime(df['visit_date'])
df['quarter'] = df['visit_date'].dt.quarter
```

```python
#let's say we're interested in warm season weekend days 
df['quarter'].isin([2, 3]) & (df['part_of_week'] == 'Weekend')
```

```python
conditions = [
    df['quarter'].isin([2, 3]) & (df['part_of_week'] == 'Weekend'),
    df['quarter'].isin([2, 3]) & (df['part_of_week'] == 'Weekday'),
    df['quarter'].isin([1, 4]) & (df['part_of_week'] == 'Weekend'),
    df['quarter'].isin([1, 4]) & (df['part_of_week'] == 'Weekday')
]
results    = [
    'Weekend Warm',
    'Weekday Warm',
    'Weekend Cold',
    'Weekday Cold'
]

df['WeatherPart'] = np.select(conditions, results, 'Other')
#arguments (conditions, results, else value)

#similar to if, elif, elif, elif, else
```

```python
df
```

Another method that is flexible, but slow, is apply()

apply() allows you to write a regular Python function and apply it to your dataset. Sadly it is about 1000 times slower than other np methods that do similar things. 

Apply is a bit of a hammer, but it can be useful for doing something very specific 

```python
#let's write a Python function! 

def clean_area_text(area_text):
    return area_text.replace('-', ' ').lower()
```

```python
df['area'].apply(clean_area_text)
#you pass the function into the apply function 
```

## Plotly! 

```python
import plotly.express as px
```

```python
dir(px)
```

```python
total_visits = df.groupby('visit_date')[['visitors']].sum()
#the double brackets turns what would be a Series into a Dataframe :) 
#sort of a shortcut for .toframe()
```

```python
#'pop out' the index
total_visits = total_visits.reset_index()

#why is this important? 
#when you do a groupby(), the "grouped_by" field gets added to the index
#resetting the index pops it out and means you can use it inside plotly
#lets you use the "grouped_by" field in the plotly arguments 
```

```python
total_visits
```

```python
px.line(total_visits, x = 'visit_date', y = 'visitors')
```

```python
#for when you need to get serious 
#grammar of graphics babyyyy
import plotly.graph_objects as go
```

```python
fig = go.Figure()
#go figure! 
```

```python
fig
```

```python
#adding a line element 
#reminiscient of R geoms
fig.add_trace(go.Scatter(x=total_visits['visit_date'], y = total_visits['visitors']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=total_visits['visit_date'], y = total_visits['visitors'], mode='markers'))
```

The `mode=` argument for scatter tells figure how to draw the scatter

```python
#let's add more to this figure 
total_reservations = df.groupby('visit_date')['reserve_visitors'].sum().to_frame()
```

### The three most common ways of encoding categorical data 

1. Ordinal Encoding 

```python
#good for tree-based models and not much else 
#not particularly sophisticated
#if something has a natural rank, then ordinal encoders can also work well 
## big, medium, small -> can specify which values map to which encodings
### can use the (mapping =) argument to get this done - list of nested dictionaries
ore = ce.OrdinalEncoder()
ore.fit_transform(df)

#takes every category value and just assigns it a value that increases 
```

2. One-hot encoding
* This is probably the standard way that people encode data 
* There are a few pathologies 
    * Probably best to only one-hot encode __some__ of your data
    * Not great for tree-based models, esp. if you have high-cardinality data 

```python
oh = ce.OneHotEncoder()
```

```python
oh.fit_transform(df[:10000])
```

3. Target Encoding
* if you have a unique ID, your TE antenna should go up 

```python
te = ce.TargetEncoder()
```

```python
df.groupby('id').mean()['visitors']
```

```python
#projecting to the entire dataset 

df.groupby('id')['visitors'].transform('mean')
```

The above is basically what target encoding does

```python
X = df.drop('visitors', axis = 1)
y = df['visitors']
```

## First Machine Learning Discussion

```python
#scikit learn is less "flat" than pandas and numpy
#we'll typically import from sub-modules, as below

from sklearn.tree import DecisionTreeRegressor
```

```python
#intialize algorithm
#sklearn is very object-oriented
tree = DecisionTreeRegressor(max_depth = 3)
```

```python
#define inputs and outputs - typical for scikit learn 
X = df.iloc[:,:-1] #grab all rows and every column except the last one 
Y = df['PRICE']
```

```python
#DO THE THING - do the pattern-matching
tree.fit(X,Y)
```

```python
#but wait, I want to see it 
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
#we don't love matplotlib, but sklearn built it's stuff on top 
```

```python
#initialize figure 
plt.figure(figsize = (16,10))
plot_tree(tree, filled=True, fontsize = 13, feature_names = X.columns);
```
